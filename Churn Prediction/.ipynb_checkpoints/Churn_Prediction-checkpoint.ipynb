## From scratch implementation

```python
class LogisticRegression_:
    def __init__(self, alpha=.1, prob_threshold=.5):
        self.alpha = alpha
        self.prob_threshold = prob_threshold

    # prepare data
    def prepareData(self, X, bias):
        if bias == True:
            weights = np.zeros((X.shape[1] + 1, 1))
            X = np.c_[np.ones((X.shape[0] , 1)), X]
        else:
            weights = np.zeros((X.shape[1], 1))
        return weights, X
            
    # calculate hx
    def sigmoid(self, z):
        sig = 1/(1 + np.exp(-z))
        return sig
    
    # fitting the mode
    def fit(self, X, Y, bias=True, max_iter=1000):
        weights, X = self.prepareData(X, bias)
        for i in range(max_iter):
            z = X @ weights
            hx = self.sigmoid(z)
            weights = weights - self.alpha * (X.T @ (hx - Y)) / len(Y)
        return weights
    
    # prediction
    def predict(self, X,  weights, bias=True):
        w, X = self.prepareData(X, bias)
        t = X @ weights
        sig = self.sigmoid(t)
        predictions = np.where(sig > self.prob_threshold, 1, 0) 
        return predictions
                
```

## Lagrange equations

## Lagrange Dual form
(**PRIMAL**)
\begin{equation}
\min_{w\  \in\  R^d,\  b\  \in\  R} ||w||^2 \\ s.t.:\  y^{(i)} (w.x^{(i)} + b) \geq 1 \ \forall i = 1, 2,..,n
\end{equation}
\
(**DUAL**)
\begin{equation}
\max_{\alpha \in R^n} \sum_{i=1}^n \alpha - \frac{1}{2} \sum_{i, j=1}^n \alpha_i \alpha_j y^{(i)} y^{(j)} (x^{(i)}.x^{(j)}) \\ s.t.: \ \sum_{i=1}^n \alpha_i y^{(i)} = 0, \\
\alpha \geq 0
\end{equation}



## **cvxopt** required form
\begin{equation}
\min{\frac{1}{2} x^T P x + q^T x}\\
s.t.: Gx \leq h \\
Ax = b
\end{equation}

*   **P** := a Matrix H of size m x m
*   **q** := $ \overrightarrow {-1} $ of size m x 1
*   **G** := a diagonal matrix of -1s of size m x m
*   **h** := $ \overrightarrow {0} $ of size m x 1
*  **A** := y the label vector of size m x 1
*   **b** := 0 a scalar

1. First let $ H_{i,j} = y^{(i)} y^{(j)} < x^{(i)}x^{(j)}>$ then the optmization becomes \\
\begin{equation}
\max_{\alpha} \sum_{i}^m \alpha_i- \frac{1}{2} \alpha^T H \alpha\\
s.t.:\  \alpha_i \geq 0, \\
\sum_{i}^m \alpha_i y^{(i)} = 0
\end{equation}


2. second convert from max to min by multiplying by -1.
\begin{equation}
\min_{\alpha} \sum_{i}^m \frac{1}{2} \alpha^T H \alpha -1^T \alpha\\
s.t.:\  -\alpha_i \leq 0, \\
y^T \alpha = 0
\end{equation}


If we have 2 data points with 2 features then \\
$X = \begin{bmatrix} x_1^{(1)} & x_2^{(1)} \\ x_1^{(2)} & x_2^{(2)} \end{bmatrix} $,  $y = \begin{bmatrix} y^{(1)} \\ y^{(2)} \end{bmatrix} $ \\

$ X * y$ in numpy will broadcast to result in  \\
$X' = \begin{bmatrix} x_1^{(1)}y^{(1)} & x_2^{(1)}y^{(1)} \\ x_1^{(2)}y^{(2)} & x_2^{(2)}y^{(2)} \end{bmatrix} $ \\
$ H = X'X^T $

## Emplementations

```python
x_dash = x * y
```

```python
H = x_dash @ x_dash.T
```

```python
H.dtype
```

```python
y.dtype
```

```python
m, n = H.shape
```

```python
p = cvxopt_matrix(H)
q = cvxopt_matrix(-np.ones((m, 1)))
G = cvxopt_matrix(-np.eye(m))
h = cvxopt_matrix(np.zeros(m))
A = cvxopt_matrix(y.reshape(1, -1))
b = cvxopt_matrix(np.zeros(1))
```

```python
sol = cvxopt_solvers.qp(p, q, G, h, A, b)
```

```python
alphas = np.array(sol['x'])
```

```python
w = (y * alphas).T @ x
```

```python
x.shape
```

```python
w = w.reshape(-1, 1)
```

```python
# selectinng support vector
s = (alphas > 1e-4).flatten()
```

```python
b = y[s] - x[s] @ w
```

```python
b.mean()
```

```python
svm_pred = (x @ w + b.mean()) * y
```

```python
final_svm_pred = (svm_pred >=0).astype(int)
```
