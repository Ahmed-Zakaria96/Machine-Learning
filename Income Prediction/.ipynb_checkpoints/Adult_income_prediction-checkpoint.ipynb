```python
# no order pattern features
nominal_cats = ['occupation', 'relationship', 'marital.status', 'workclass', 'sex']
OHE = OneHotEncoder(sparse=False).fit(Adult.train_data[nominal_cats])
ohe_train = OHE.transform(Adult.train_data[nominal_cats])
ohe_test = OHE.transform(Adult.test_data[nominal_cats])
```

```python
# has order pattern features
ordinal_cats = ['education']
fill = {
    "Preschool": 1,
    "1st-4th": 2,
    "5th-6th": 3,
    "7th-8th": 4,
    "9th": 5,
    "10th": 6,
    "11th": 7,
    "12th": 8,
    "HS-grad": 9,
    "Prof-school": 10,
    "Assoc-acdm": 11,
    "Assoc-voc": 12,
    "Some-college": 13,
    "Bachelors": 14,
    "Masters": 15,
    "Doctorate": 16,
}
ord_train = Adult.train_data['education'].replace(fill).values.reshape((-1, 1))
ord_test = Adult.test_data['education'].replace(fill).values.reshape((-1, 1))
```

```python
encoded_train = np.c_[Adult.train_data[Adult.grabNumeric(True)], ohe_train, ord_train]
cols = Adult.grabNumeric(True) + list(OHE.get_feature_names_out()) + ['education']
final_train = pd.DataFrame(encoded_train, columns=cols )
```

```python
encoded_test = np.c_[Adult.test_data[Adult.grabNumeric(True)], ohe_test, ord_test]
cols = Adult.grabNumeric(True) + list(OHE.get_feature_names_out()) + ['education']
final_test = pd.DataFrame(encoded_test, columns=cols)
```

> ## 3.7- Solving the problem of imbalance

```python
final_train['income'].value_counts()
```

```python
len(final_train[final_train['income'] == 1])/ len(final_train)
```

```python
rare_class = final_train[final_train['income'] == 1]
dominant_class= final_train[final_train['income'] == 0]

under_sampled = dominant_class.sample(frac = 0.3, replace = False, random_state = 42)
train_undersampled = pd.concat((under_sampled, rare_class), axis = 0)
```

```python
# final_train_data, final_val_data
train_x = train_undersampled.drop(columns =['income'])
train_y = train_undersampled[['income']]

val_x = final_test.drop(columns=['income'])
val_y = final_test[['income']]
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
```

```python
p = np.arange(1,15)
```

```python
results = []
for i in p:
    KNN = KNeighborsClassifier(metric='minkowski', p=i)
    KNN.fit(xTrain, train_y)
    y_ = KNN.predict(xVal)
    score = f1_score(val_y, y_)
    results.append({i: score}) 
```

```python
results
```

```python
neighbors_results = []
for i in range(1, 51, 5):
    KNN = KNeighborsClassifier(n_neighbors=i, metric='minkowski', p=5)
    KNN.fit(xTrain, train_y)
    y_ = KNN.predict(xVal)
    score = f1_score(val_y, y_)
    neighbors_results.append({i: score}) 
```

```python
neighbors_results
```

```python
weights = ['uniform', 'distance']
weights_score = []
for i in range(2):
    KNN = KNeighborsClassifier(weights=weights[i], n_neighbors=16, metric='minkowski', p=5)
    KNN.fit(xTrain, train_y)
    y_ = KNN.predict(xVal)
    score = f1_score(val_y, y_)
    weights_score.append({weights[i]: score}) 
```

## Tunning hyper parameters

```python
sns.set_theme()
```

```python
depth_results = []
depth = np.arange(3, 30)
for i in depth:
    clf = tree.DecisionTreeClassifier(max_depth=i)
    clf = clf.fit(train_x, train_y)
    train_pred_y = clf.predict(train_x)
    train_score = f1_score(train_y, train_pred_y)
    test_pred_y = clf.predict(val_x)
    test_score = f1_score(val_y, test_pred_y)
    r = {'train_score': train_score, 'val_score': test_score}
    depth_results.append({f"{i}": r})
```

```python
train_d = [x[list(x.keys())[0]]['train_score'] for x in depth_results]
val_d = [x[list(x.keys())[0]]['val_score'] for x in depth_results]
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot()
sns.lineplot(x=depth, y=train_d, ax=ax)
sns.lineplot(x=depth, y=val_d, ax=ax)
ax.set_xlabel("Depth")
ax.set_ylabel("Score")
plt.legend(labels=['Train Score', 'Validation Score']);
```

```python
min_samples_split_results = []
min_samples_split = np.arange(2, 400, 2)
for i in range(2, 400, 2):
    clf = tree.DecisionTreeClassifier(min_samples_split=i)
    clf = clf.fit(train_x, train_y)
    train_pred_y = clf.predict(train_x)
    train_score = f1_score(train_y, train_pred_y)
    test_pred_y = clf.predict(val_x)
    test_score = f1_score(val_y, test_pred_y)
    r = {'train_score': train_score, 'val_score': test_score}
    min_samples_split_results.append({i: r})
```

```python
samples_train_d = [x[list(x.keys())[0]]['train_score'] for x in min_samples_split_results]
samples_val_d = [x[list(x.keys())[0]]['val_score'] for x in min_samples_split_results]
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot()
sns.lineplot(x=min_samples_split, y=samples_train_d, ax=ax)
sns.lineplot(x=min_samples_split, y=samples_val_d, ax=ax)
ax.set_xlabel("min_samples_split")
ax.set_ylabel("Score")
plt.legend(labels=['Train Score', 'Validation Score']);
```

```python
min_samples_leaf_results = []
min_samples_leaf = np.arange(1, 50)
for i in min_samples_leaf:
    clf = tree.DecisionTreeClassifier(min_samples_leaf=i)
    clf = clf.fit(train_x, train_y)
    train_pred_y = clf.predict(train_x)
    train_score = f1_score(train_y, train_pred_y)
    test_pred_y = clf.predict(val_x)
    test_score = f1_score(val_y, test_pred_y)
    r = {'train_score': train_score, 'val_score': test_score}
    min_samples_leaf_results.append({i: r})
```

```python
leaf_train_d = [x[list(x.keys())[0]]['train_score'] for x in min_samples_leaf_results]
leaf_val_d = [x[list(x.keys())[0]]['val_score'] for x in min_samples_leaf_results]
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot()
sns.lineplot(x=min_samples_leaf, y=leaf_train_d, ax=ax)
sns.lineplot(x=min_samples_leaf, y=leaf_val_d, ax=ax)
ax.set_xlabel("min_samples_leaf")
ax.set_ylabel("Score")
plt.legend(labels=['Train Score', 'Validation Score']);
```

```python
max_features_results = []
max_features = np.arange(1, 42)
for i in max_features:
    clf = tree.DecisionTreeClassifier(max_features=i)
    clf = clf.fit(train_x, train_y)
    train_pred_y = clf.predict(train_x)
    train_score = f1_score(train_y, train_pred_y)
    test_pred_y = clf.predict(val_x)
    test_score = f1_score(val_y, test_pred_y)
    r = {'train_score': train_score, 'val_score': test_score}
    max_features_results.append({i: r})
```

```python
maxFeat_train_d = [x[list(x.keys())[0]]['train_score'] for x in max_features_results]
maxFeat_val_d = [x[list(x.keys())[0]]['val_score'] for x in max_features_results]
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot()
sns.lineplot(x=max_features, y=maxFeat_train_d, ax=ax)
sns.lineplot(x=max_features, y=maxFeat_val_d, ax=ax)
ax.set_xlabel("max_features")
ax.set_ylabel("Score")
plt.legend(labels=['Train Score', 'Validation Score']);
```

```python
max_leaf_nodes_results = []
max_leaf_nodes = np.arange(2, 100)
for i in max_leaf_nodes:
    clf = tree.DecisionTreeClassifier(max_leaf_nodes=i)
    clf = clf.fit(train_x, train_y)
    train_pred_y = clf.predict(train_x)
    train_score = f1_score(train_y, train_pred_y)
    test_pred_y = clf.predict(val_x)
    test_score = f1_score(val_y, test_pred_y)
    r = {'train_score': train_score, 'val_score': test_score}
    max_leaf_nodes_results.append({i: r})
```

```python
leafNode_train_d = [x[list(x.keys())[0]]['train_score'] for x in max_leaf_nodes_results]
leafNode_val_d = [x[list(x.keys())[0]]['val_score'] for x in max_leaf_nodes_results]
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot()
sns.lineplot(x=max_leaf_nodes, y=leafNode_train_d, ax=ax)
sns.lineplot(x=max_leaf_nodes, y=leafNode_val_d, ax=ax)
ax.set_xlabel("max_leaf_nodes")
ax.set_ylabel("Score")
plt.legend(labels=['Train Score', 'Validation Score']);
```

```python
ccp_alpha_results = []
ccp_alpha = np.arange(.001, .1, .001)
for i in ccp_alpha:
    clf = tree.DecisionTreeClassifier(ccp_alpha=i)
    clf = clf.fit(train_x, train_y)
    train_pred_y = clf.predict(train_x)
    train_score = f1_score(train_y, train_pred_y)
    test_pred_y = clf.predict(val_x)
    test_score = f1_score(val_y, test_pred_y)
    r = {'train_score': train_score, 'val_score': test_score}
    ccp_alpha_results.append({i: r})
```

```python
alpha_train_d = [x[list(x.keys())[0]]['train_score'] for x in ccp_alpha_results]
alpha_val_d = [x[list(x.keys())[0]]['val_score'] for x in ccp_alpha_results]
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot()
sns.lineplot(x=ccp_alpha, y=alpha_train_d, ax=ax)
sns.lineplot(x=ccp_alpha, y=alpha_val_d, ax=ax)
ax.set_xlabel("ccp_alpha")
ax.set_ylabel("Score")
plt.legend(labels=['Train Score', 'Validation Score']);
```
